```python
data = [10, 20, 30, 40, 50]
n = len(data)
mean = sum(data) / n # Mean value computed
squared_diffs = [(x-mean) ** 2 for x in data] #Squared differences
variance =  sum(squared_diffs) / n
standard_deviation = variance ** 0.5

print(f"Mean: {mean}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {standard_deviation}")
```
