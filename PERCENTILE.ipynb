```python
import numpy as np
data = [15, 20, 35, 40, 50]

#Q1(25th), Q2(50th), and Q3(75th) percentiles
percentile_25 = np.percentile(data, 25)
percentile_50 = np.percentile(data, 50) #Q2 also known as Median
percentile_75 = np.percentile(data, 75)

print("25th Percentile (Q1): ", percentile_25 )
print("50th Percentile (Q2/Median): ", percentile_50)
print("75th Percentile (Q3): ", percentile_75)
```
